```c++
template <size_t N>
constexpr auto extract_bits(int x) {
    if constexpr (N == 0) return tuple{};
    else return tuple_cat(tuple{static_cast<bool>(x & 1)}, extract_bits<N-1>(x >> 1));
}
```
